```python
import pandas as pd
url: str = '/Users/jjmtz/Desktop/automobileEDA.csv'
df: pd.DataFrame = pd.read_csv(url)
df
```

```python
df.info()
```

```python
from sklearn.linear_model import LinearRegression
X = df[['horsepower']]
Y = df['price']
lr = LinearRegression()
lr.fit(X,Y)
yhat = lr.predict(X)

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('dark_background')

plt.figure(figsize=(8,6))
plt.scatter(df['horsepower'], df['price'])
plt.scatter(df['horsepower'], yhat)
```

```python
import seaborn as sns
sns.distplot(yhat, hist = False)
sns.distplot(df['price'], hist = False)
```

```python
x = df['horsepower']
y = df['price']
dist_range = np.linspace(x.min(), x.max(), len(x))
yhat = lr.intercept_ + lr.coef_[0] * df['horsepower']
plt.plot(x, yhat)
plt.scatter(x,y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

feature_cols = ['normalized-losses', 'horsepower', 'city-mpg', 'highway-mpg', 'engine-size']
X = df[feature_cols]
Y = df['price']

Input = [('scaler', StandardScaler()), ('polynomial', PolynomialFeatures(degree=4)),('model', LinearRegression())]
pipe5d5x = Pipeline(Input)
_ = pipe5d5x.fit(X, Y)

_ = plt.figure(figsize=(10,7))
_ = sns.distplot(pipe5d5x.predict(X), hist=False, color='crimson')
_ = sns.distplot(Y, hist=False, color='lightblue')
_ = plt.legend(['Predicted distribution', 'Actual distribution'])
```
